```python
#This is for building a sequential neural network
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential #A layered neural network 
from keras.layers import Dense #Layers 

import warnings
warnings.filterwarnings("ignore")

#Define the model
model = Sequential()
```

```python
#Define Dataset
health_db = pd.read_csv("health_dataset.csv")

#Transform objects into numerics
columns = ["exercise_frequency", "part_time_job", "extracurricular_participation"]
for feature in columns:
    le = LabelEncoder()
    health_db[feature] = le.fit_transform(health_db[feature])

#Drop unused columns
health_db = health_db.drop(["student_id", "age", "gender", "attendance_percentage", 
                            "parental_education_level", "internet_quality", "exam_score"])
```

```python



#Add the layers of the neural network after defining the model
#Layer format: (Dense(a, b, c, d))
#a: The dimension. Equals the number of inputs (columns used excluding target)
#b: Kernel_initializer. Sets the weights of the layer
#c: Activation.
model.add(Dense(7, kernel_intializer = "uniform", activation = "relu", input_dm = 5))

```
