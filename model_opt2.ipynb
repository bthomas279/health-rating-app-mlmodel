```python
#This is testing a sequential neural network
from keras.models import Sequential #A layered neural network 
from keras.layers import Dense #Layers 

#Define the model
model = Sequential()
```

```python
#Add the layers of the neural network after defining the model
#Layer format: (Dense(a, b, c, d))
#a: The dimension. Equals the number of inputs (columns used excluding target)
#b: Kernel_initializer. Sets the weights of the layer
#c: Activation
model.add(Dense(7, kernel_intializer = "uniform", activation = "relu", input_dm = 5))

```
